```python
#import file and extract reviews.Text column for analysis
import os 
import pandas as pd
df=pd.read_csv("E:/CPC353 Assgn/Amazon tablet dataset/Tablet_Amazon.csv")
df.rename(columns={"reviews.text":"Reviews"},inplace=True)
df.head()
```

```python
#remove all column except review.text
df=df.iloc[:,16:17]
df.head()
```

```python
#change the data type to string
df['Reviews'] = df['Reviews'].astype(str)
```

```python
#data preprocessing to remove punctuation, special characters and whitespaces
import re, string
def clean_text(text):
    text=re.sub(r"\?", " ", text)
    text=re.sub(r"\!", " ", text)
    text=re.sub(r"\:", " ", text)
    text=re.sub(r"\n", "", text)
    text=re.sub(r"\’"," ",text)
    text=re.sub(r"\“"," ",text)
    text=re.sub(r"\”"," ",text)
    text=re.sub(r"\*"," ",text)
    text=re.sub(r'([.])\1+', r'\1', text)
    rexp=r"\.(?=[A-Za-z]{1,15})"
    text=re.sub(rexp, ". ", text)
    return text
    
```

```python
# Tokenize the sentences
import nltk
from nltk import word_tokenize,sent_tokenize
df["Tokenize_Review"]=df["Reviews"].apply(lambda x:clean_text(x))
```

```python
# Change all tokenize sentences to lowercase
df["Tokenize_Review"]=df["Tokenize_Review"].apply(lambda x:word_tokenize(x.lower()))
df.head()
```

```python
# apply part of speech and WordNet tagging on the tokenize sentences
# stopwords is remove 
from nltk.corpus import wordnet
from nltk.corpus import stopwords
pos_dict={'J':wordnet.ADJ,'V':wordnet.VERB,'N':wordnet.NOUN,'R':wordnet.ADV}
def pos_tag(text):
    tags=nltk.pos_tag(text)
    stop_word=stopwords.words("english")
    newlist=[]
    for word,tag in tags:
        if(word not in set(stop_word)):
            newlist.append(tuple([word,pos_dict.get(tag[0])]))
    return newlist
```

```python
df["pos_tag"]=df["Tokenize_Review"].apply(lambda x:pos_tag(x))
df.head()
```

```python
#Lemmatize the words to its root form according to the WordNet tagging
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()
def lemmatize(text):
    lemma_rew=""
    for word,pos in text:
        if pos:
            lemma=wordnet_lemmatizer.lemmatize(word,pos=pos)
            lemma_rew=lemma_rew + " " + lemma
        
    return lemma_rew
    
```

```python
df["lemmatize"]=df["pos_tag"].apply(lambda x:lemmatize(x))
df.head()
```

```python
#import textblob to calculate the sentiment polarity intensity
from textblob import TextBlob
def get_polarity(reviews):
    return TextBlob(reviews).sentiment.polarity
# get the sentiment orientation based on the sentiment polarity
def get_analysis(score):
    if score<0:
        return "negative"
    elif score==0:
        return "neutral"
    else:
        return "positive"
```

```python
df_new=pd.DataFrame(df[['Reviews','lemmatize']])
df_new.head()
```

```python
df_new['Polarity']=df_new['lemmatize'].apply(get_polarity)
df_new['Analysis']=df_new['Polarity'].apply(get_analysis)
df_new.head()
```

```python
# Total length of positive,neutral and negative based on the lemmatized sentences
from collections import Counter
print(Counter(df_new['Analysis']))
```

```python
# Perform unigram
unigram=pd.DataFrame(Counter(" ".join(df_new['lemmatize']).split()).most_common(100),columns=['unigram','frequency'])
```

```python
# Perform sentiment polarity and orientation on each unigram
unigram['Polarity']=unigram['unigram'].apply(get_polarity)
unigram['analysis']=unigram['Polarity'].apply(get_analysis)
```

```python
unigram.head(10)
```

```python
# Total length of positive,neutral and negative for unigram
Counter(unigram['analysis'])
```

```python
# Perform bigram
from nltk.collocations import *
bigram_measures = nltk.collocations.BigramAssocMeasures()
finder = BigramCollocationFinder.from_words(" ".join(df_new['lemmatize']).split(),window_size=5)
#keeping bigrams that occur at least 5 times
finder.apply_freq_filter(5)
bigram_list_tuples=finder.nbest(bigram_measures.likelihood_ratio, 50)
frequency = finder.ngram_fd.items()
```

```python
# sort bigram according to highest frequency
frequency_sorted=sorted(finder.ngram_fd.items(), key=lambda x: x[1], reverse=True)
bigram_extraction=pd.DataFrame(frequency_sorted,columns=['Bigram','Frequency'])
```

```python
# create a new variable called 'new' to form a new bigram 
bigram_extraction['new']=bigram_extraction['Bigram'].apply(lambda x:" ".join(x))
```

```python
# Perform sentiment polarity and orientation on each bigram
bigram_extraction['polarity']=bigram_extraction['new'].apply(get_polarity)
bigram_extraction['analysis']=bigram_extraction['polarity'].apply(get_analysis)
bigram_extraction.head(10)
```

```python
# Analyse the sentiment of several features of tablet
feature=bigram_extraction[bigram_extraction['new'].str.contains('screen|battery|memory|service')]
bigram_feature=feature.sort_values(['polarity','Frequency'],ascending=False)
bigram_feature.head(10)
```

```python
# Total length of positive,neutral and negative based on the bigram
print(Counter(bigram_extraction['analysis']))
print(len(bigram_extraction))
```

```python
# Perform Trigram
from nltk.collocations import *
trigram_measures = nltk.collocations.TrigramAssocMeasures()
finder = TrigramCollocationFinder.from_words(" ".join(df_new['lemmatize']).split(),window_size=5)
#keeping trigrams that occur at least 5 times
finder.apply_freq_filter(5)
trigram_list_tuples=finder.nbest(trigram_measures.likelihood_ratio, 50)
frequency = finder.ngram_fd.items()
```

```python
# sort trigram according to highest frequency
frequency_sorted=sorted(finder.ngram_fd.items(), key=lambda x: x[1], reverse=True)
trigram_extraction=pd.DataFrame(frequency_sorted,columns=['Trigram','Frequency'])
```

```python
trigram_extraction['new']=trigram_extraction['Trigram'].apply(lambda x:" ".join(x))
```

```python
# Perform sentiment polarity and orientation on each trigram
trigram_extraction['polarity']=trigram_extraction['new'].apply(get_polarity)
trigram_extraction['analysis']=trigram_extraction['polarity'].apply(get_analysis)
trigram_extraction.head(10)
```

```python
# Analyse the sentiment of several features of tablet
feature=trigram_extraction[trigram_extraction['new'].str.contains('screen|battery|memory|service')]
trigram_feature=feature.sort_values(['polarity','Frequency'],ascending=False)
trigram_feature.head(10)
```

```python
# Total length of positive,neutral and negative based on the trigram
print(Counter(trigram_extraction['analysis']))
print(len(trigram_extraction))
```
